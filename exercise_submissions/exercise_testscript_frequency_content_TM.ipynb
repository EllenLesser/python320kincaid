```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
import scipy.fftpack
'''
Add any packages that you might need 
'''
```

```python
sample_rate, bird_signal_int = scipy.io.wavfile.read("data//birds001.wav")
bird_signal_float = np.array(bird_signal_int,dtype=float)
```

```python
def find_frequency_spectrum(signal,sampling_rate):
     N = len(signal)
     yf =scipy.fftpack.fft(signal)
     print( yf.shape ) 
     xf = np.arange(0.0,N//2,1) /N*sampling_rate
     yf_norm = 2.0/N*np.abs(yf[:N//2])
     return xf,yf_norm
```

```python
t = np.arange(0,len(bird_signal_float)/sample_rate,1/sample_rate)
freq,amp = find_frequency_spectrum( bird_signal_float ,sample_rate)

plt.figure()
plt.subplot(211)
plt.plot(t,bird_signal_float)
plt.subplot(212)
plt.plot(freq,amp)
plt.xlim( freq[[0,amp.argsort()[-2]]]*2)
plt.show()
```

```python
sample_rate_sine = 1000
t_sine = np.arange(0,10,1/sample_rate_sine)
y_sine = 3*np.sin(2*np.pi*t_sine)+ 0.5*np.sin(2*np.pi* 3.5*t_sine)
freq_sine,amp_sine = find_frequency_spectrum(y_sine,sample_rate_sine)

plt.figure()
plt.subplot(211)
plt.plot(t_sine,y_sine)
plt.subplot(212)
plt.plot(freq_sine,amp_sine)
plt.xlim( freq_sine[[0,amp_sine.argsort()[-2]]]*2)
plt.show()
```
