```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from chorogrid import Colorbin, Chorogrid
```

```python
df = pd.read_csv('data/Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv')
```

```python
print( df.index )
print( df.columns)
```

```python
Topics = df.Topic.unique()
print(Topics)

FruitsVeg = df[ df.Topic == Topics[1]]
FruitVegQ = FruitsVeg.Question.unique()
```

```python
Fruits = FruitsVeg[FruitsVeg.Question== FruitVegQ[0]]
```

```python
FruitVegQ[0]
```

```python
 Fruits.LocationDesc.unique() 
```

```python
FruitsWA = Fruits[Fruits.LocationDesc == 'Washington']
FruitsWA.Education.unique()
```

```python
fruitsWAgrads = FruitsWA[FruitsWA.Education == 'College graduate']
```

```python
fruitsWAgradData = fruitsWAgrads.loc[:,'Data_Value_Unit':'Gender']
```

```python
plt.figure()

low_error =  fruitsWAgradData.Data_Value - fruitsWAgradData.Low_Confidence_Limit.values
high_error = fruitsWAgradData['High_Confidence_Limit '].values -  fruitsWAgradData.Data_Value

years = fruitsWAgrads['YearStart']
plt.errorbar(years, fruitsWAgradData.Data_Value ,       yerr = [ low_error, high_error ],  fmt='o')

ax = plt.gca()
ax.set_xticks( years)
ax.set_xlabel('year of survey')
ax.set_ylabel('% of college grads with less than 1 fruit consumption a day')

plt.show()
```

```python
mask = (df.LocationDesc == 'Washington') \
        & (df.Education == 'College graduate') \
        & (df.Question == 'Percent of adults who report consuming fruit less than one time daily')
df.loc[mask,['Data_Value','Low_Confidence_Limit','High_Confidence_Limit '] ]
```

```python
dfSub = df[ (df.YearStart==2011) & (df.StratificationID1=='OVERALL') & (df.QuestionID == 'Q036') & (df.LocationAbbr != 'US')]

dfSub[ ['LocationDesc','Data_Value'] ].sort_values(by=['Data_Value'])
```

```python


myColorBrew = ['#d53e4f','#f46d43','#fdae61','#fee08b','#ffffbf','#e6f598','#abdda4','#66c2a5','#3288bd']
# colorbrew link: http://colorbrewer2.org/#type=diverging&scheme=Spectral&n=9

colorData = dfSub.Data_Value
states = dfSub.LocationAbbr

mybin = Colorbin(colorData , myColorBrew, proportional=True, decimals=None)
mybin.set_decimals(1)
mybin.recalc(fenceposts=True)
mybin.calc_complements(0.5, '#e0e0e0', '#101010')
colors_by_state = mybin.colors_out

cg = Chorogrid('data/usa_states.csv', states, colors_by_state)
cg.draw_map(spacing_dict={'legend_offset': [-150,-25]})
cg.set_title(dfSub.Question.iloc[0], font_dict={'font-size': 19})

legend_colors = mybin.colors_in
legend_labels = mybin.labels
cg.set_legend(legend_colors, legend_labels, title='')
cg.draw_map(spacing_dict={'legend_offset':[-300,-200],'margin_top': 30}) # otherwise legend will be cut o
cg.done(show=True)
```
