```python
def find_frequency_spectrum(signal,sampling_rate):
     N = len(signal)
     yf =scipy.fftpack.fft(signal)
     xf = np.arange(0.0,N//2,1) /N*sampling_rate
     yf_norm = 2.0/N*np.abs(yf[:N//2])
     return xf,yf_norm
```
