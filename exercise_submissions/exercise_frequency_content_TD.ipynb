```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make an arbitary sine wave to do fft on
duration=10
Rs1 = 100.0
Ts1 = 1.0/Rs1
freq1=7
freq2=13


t1 = np.arange(0,duration,Ts1)
y1 = 5* np.sin(2 * np.pi * freq1 * t1 + np.pi/6) + 8* np.sin(2* np.pi * freq2 * t1)
```

```python
y1.shape
```

```python
# check what the wave looks like
plt.plot(t1, y1)
plt.xlabel('time(sec)')
plt.ylabel('Amplitude(arbitary)')
plt.show()
```

```python
signal_fft=np.fft.fft(y1)

# get just half the signal that goes from [0 Nyquist/2] 
half_len=round(len(signal_fft)/2)
Half_signal_fft=signal_fft[0:half_len]

# normalize the amplitude
shape=y1.shape
normalize_vector=np.ones(shape)
normalize_fft= np.fft.fft(normalize_vector)
Half_normalize_fft=normalize_fft[0:half_len]
```

```python
freq_axis= np.arange(0,Rs1/2,1/duration)
freq_axis.shape
```

```python
fig, axes = plt.subplots(ncols = 2, sharey=True)
axes[0].plot(freq_axis, abs(Half_signal_fft))
axes[0].set_title('SineWave')

axes[1].plot(freq_axis, abs(Half_normalize_fft))
axes[1].set_title('NormalizationWave')

plt.show()
```

```python
amp=max(abs(Half_normalize_fft))
normalized_amp_fft= abs(Half_signal_fft)*2/amp
plt.plot(freq_axis,normalized_amp_fft)
plt.title('Final FFT curve')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Amplitude(V)')
plt.show()
```
