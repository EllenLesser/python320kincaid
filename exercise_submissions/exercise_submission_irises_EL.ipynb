```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as pch

%matplotlib inline
plt.style.use('./data/EL_figures.mplstyle')
```

```python
def plot_all_df(df):
    '''Plot all variables against each other for any Pandas dataframe'''
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    cols = len(df.columns)
    fig = plt.figure(figsize=(cols*5, cols*5))
    
    for n in range(0, cols):
        for a in range(0, cols): 
            name, name2 = df.columns[n], df.columns[a]
            ax = fig.add_subplot(cols, cols, n*cols+(a+1)) 
            # add 1 because matplotlib is not 0-indexed for matlab compatibility
            if name != name2:
                val, val2 = df[name], df[name2]
                ax.set_xlabel(name)
                ax.set_ylabel(name2)

                if df[name].dtype in numerics and df[name2].dtype in numerics:
                    ax.scatter(val, val2)
                elif df[name].dtype in numerics:
                    labels = list(set(df[name2].tolist()))
                    data = [df[df[name2] == x][name].tolist() for x in labels]
                    ax.boxplot(data, vert=False)
                    ax.set_yticklabels(labels)
                elif df[name2].dtype in numerics:
                    labels = list(set(df[name].tolist()))
                    data = [df[df[name] == x][name2].tolist() for x in labels]
                    ax.boxplot(data)
                    ax.set_xticklabels(labels)
            else: 
                ax.add_patch(pch.Rectangle((0, 0), 100, 100, fc='#F0F0F0'))
                ax.set_ylim(10, 20)
                ax.set_xlim(10, 20)
                ax.axis('off')
                
    plt.show()
    plt.clf()
    plt.close('all')
```

```python
''' EXERCISE ANSWERS PARTS 1~2 '''

# Read data as Pandas dataframe
df = pd.read_csv('./data/iris.csv')

# Convert data to np 
npdf = df.as_matrix()

# Plot all variables against each other to see trends
plot_all_df(df)
```

```python
''' EXERCISE ANSWER PART 3 '''

from sklearn.decomposition import PCA as sklearnPCA

df = pd.read_csv('./data/iris.csv')
y = df['variety'] # Split off classifications
X = df.drop('variety', axis=1) # Split off features
X_norm = (X - X.min())/(X.max() - X.min()) # feature scaling

pca = sklearnPCA(n_components=2) # 2-dimensional PCA
transformed = pd.DataFrame(pca.fit_transform(X_norm))

# Map colors on labels
transformed['variety'] = y
dfcolor = pd.DataFrame([['Setosa','red'],['Versicolor','blue'],['Virginica','yellow']],
                        columns=['variety','color'])
mergeddf = pd.merge(transformed, dfcolor)
plt.scatter(mergeddf[0], mergeddf[1], c=mergeddf['color'])
plt.show()
```
